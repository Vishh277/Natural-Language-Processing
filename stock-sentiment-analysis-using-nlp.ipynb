```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/kaggle/input/nlp-basic/Stock-Sentiment-Analysis-master/Data.csv',encoding='ISO-8859-1')
df.head()
```

#### Target : Label : 1 - Stock increaase , 0- Stock Decrease
#### Date ranges from 2008 - 2016

## Predict Stock Price based on Sentiment Analysis


```python
train = df[df['Date'] < '20150101'] #  Data till 2015
test = df[df['Date'] > '20141231'] # Data after 2015
```

#### Data Pre-processing

```python
## Have only text data - No punctuations, No numbers, No special characters

data = train.iloc[:,2:27] # Select only the Features
data.replace("[^a-zA-Z]"," ",regex=True,inplace=True)

# Rename Column Names

list1 = [i for i in range(25)]
new_index = [str(i) for i in list1]

data.columns = new_index
data.head()
```

```python
## Convert to lower case - because same words with small and capital words will be considered as small letters

for i in new_index:
    data[i] = data[i].str.lower()
data.head()
```

```python
' '.join(str(x) for x in data.iloc[1,:])
```

```python
#Now we have to consider all the headlines of a single day as a paragraph

headlines = []
for row in range(len(data.index)):
    headlines.append(' '.join(str(x) for x in data.iloc[row,0:25]))
```

## NLP Implementation

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.ensemble import RandomForestClassifier
```

#### BOW

```python
countvector = CountVectorizer(ngram_range=(2,2))  # Bigram
train_data = countvector.fit_transform(headlines)
```

#### Random Classifier

```python
randomclassifer = RandomForestClassifier(n_estimators=200,criterion='entropy')
randomclassifer.fit(train_data,train['Label'])
```

#### Predict Test Label

We have only converted Train data into paragraph, now we have to do it for test dataset

```python
test.head()
```

```python
# Convert into a paragraph
test_transform = []
for row in range(len(test.index)):
    test_transform.append(' '.join(str(x) for x in test.iloc[row,2:27]))


test_data = countvector.transform(test_transform) 
predictions = randomclassifer.predict(test_data)    
```

### Evaluating Metrics

```python
from sklearn.metrics import classification_report,confusion_matrix, accuracy_score
```

```python
matrix = confusion_matrix(test['Label'],predictions)
print('Confusion Matrix',matrix)
score = accuracy_score(test['Label'],predictions)
print('Accuracy',score)
report = classification_report(test['Label'],predictions)
print('Class Report',report)
```
